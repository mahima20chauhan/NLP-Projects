```
import pandas as pd
```

```
df = pd.read_csv("/content/movies.csv")
print("First 5 rows of dataset:")
print(df.head())
```

```
df = df[['movie_name', 'year_of_release', 'imdb_rating']]
print("\nSelected columns:")
print(df.head())
```

```
df = df.rename(columns={
    'movie_name': 'Movie',
    'year_of_release': 'Year',
    'imdb_rating': 'Rating'
})
print("\nRenamed columns:")
print(df.head())
```

```
df['Rating'] = pd.to_numeric(df['Rating'], errors='coerce')
```

```
df['Year'] = pd.to_numeric(df['Year'].str.replace(r'[()]', '', regex=True), errors='coerce')
df['Decade'] = (df['Year'] // 10) * 10
print("\nAdded 'Decade' column:")
print(df.head())
```

```
df_sorted = df.sort_values(by=['Decade', 'Rating'], ascending=[True, False])
print("\nSorted by Decade and Rating:")
print(df_sorted.head(10))
```

```
top5 = df.groupby('Decade').head(5)
```

```
print("Top 5 Movies per Decade\n")
print(top5.to_string(index=False))
```
