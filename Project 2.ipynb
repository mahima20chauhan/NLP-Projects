```python
import pandas as pd
```

```python
df = pd.read_csv("/content/Flipcart_mobile1.csv")
print("First 5 rows of dataset:")
print(df.head())
```

```python
df = df[["Name", "Discounted_Price", "Rating"]]
print("\nSelected important columns:")
print(df.head())
```

```python
df["Discounted_Price"] = df["Discounted_Price"].replace({"₹": "", ",": ""}, regex=True).astype(float)
print("\nCleaned 'Price' column:")
print(df.head())
```

```python
df["Rating"] = pd.to_numeric(df["Rating"], errors="coerce")
print("\nCleaned 'Rating' column:")
print(df.head())
```

```python
df.to_excel("Ecommerce_Products.xlsx", index=False)
print("\nCleaned data saved to 'Ecommerce_Products.xlsx'")
```

```python
avg_price = df["Discounted_Price"].mean()
print("\nAverage Price of Products:")
print(avg_price)
```

```python
top_rated = df.loc[df["Rating"].idxmax()]
print("\nTop-rated Product:")
print(top_rated)
```
